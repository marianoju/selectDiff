```python
from bs4 import BeautifulSoup
import urllib.request
import re
import os
```

```python
# replaces HTML tags with nothing 
def strip_tags(description):
    return re.sub('<.*?>', '', description)
```

```python
# reads XML file from online source 
url = 'https://stackoverflow.com/feeds'

# parses XML document and returns a `BeautifulSoup` object (aka soup)
with urllib.request.urlopen(url) as response:
    soup = BeautifulSoup(response.read(), "xml")
```

```python
def remove_newline(text):
    return re.sub('\n', '', text)
```

```python
remove_newline(strip_tags(soup.summary.string))
```

```python
def singularize_whitespaces(text):
    return re.sub('  ', ' ', text)
```

```python
singularize_whitespaces(remove_newline(strip_tags(soup.summary.string)))
```

```python
def strip_non_ascii_chars(text):
    return re.sub(r'[^\x00-\x7F]+',' ', text)
```

```python
def sanitize_string(string):
    return ''.join([c for c in string if c.isalpha() or c.isdigit()]).rstrip() 

def mkdir(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)
```

```python
# finds and prints all entries in the soup object 
for entry in soup.find_all('entry'):
    filename = f'{sanitize_string(entry.id.string)}.txt'
    directory = sanitize_string(url)
    rootpath = '/'.join(['../data_new', directory])
    path = '/'.join([rootpath, filename])
    # + entry.published.string
    mkdir(rootpath)
    with open(path, 'w') as file:
        stripped_strings = ''
        for string in entry.summary.stripped_strings:
            stripped_strings += strip_tags(repr(string))
        file.write(stripped_strings)
```
